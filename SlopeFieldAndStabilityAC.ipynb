#### Consider the differential equation (DE) of the form $ \displaystyle\frac{dy}{dt}=f(y)g(t)$ where the RHS of the DE can be factored as a function of $x$ and $y$. Such DEs are called separable equations. Look at the slope field of a separable equation.

### First Example: Find the slope field of  $y^\prime = ky$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def slopefield(f,tinterval,yinterval,tstep,ystep):
    t = np.arange(tinterval[0],tinterval[1],tstep)
    y = np.arange(yinterval[0],yinterval[1],ystep)
    L = 0.7*min(tstep,ystep)
    for i in range(len(t)):
        for j in range(len(y)):
            slope = f(t[i],y[j])
            theta = np.arctan(slope)
            dy = L*np.sin(theta)
            dt = L*np.cos(theta)
            plt.plot([t[i],t[i] + dt],[y[j],y[j] + dy],'b')
```

### A. (k=2): Find the slope field of  $y^\prime = ky$.

```python
f1 = lambda t, y: 2*y
slopefield(f1,[0,4],[-2,2],0.2,0.2)
```

### B. (k=-2): Find the slope field of  $y^\prime = ky$.

```python
f2 = lambda t, y: -2*y
slopefield(f2,[0,4],[-2,2],0.2,0.2)
```
