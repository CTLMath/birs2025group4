## Plotting Slope Fields
Given a differential equation of the form $\frac{dy}{dt}=f(t,y)$, how can we represent graphically the information in this differential equation? One way it so use **Slope Field**. This creates a grid in the $ty-$plane, and at each gridpoint it plots a little line whose slope is given by $f(t,y)$.  

```python
def slopefield(f,tinterval,yinterval,tstep,ystep):
    t = np.arange(tinterval[0],tinterval[1],tstep)
    y = np.arange(yinterval[0],yinterval[1],ystep)
    L = 0.7*min(tstep,ystep)
    for i in range(len(t)):
        for j in range(len(y)):
            slope = f(t[i],y[j])
            theta = np.arctan(slope)
            dy = L*np.sin(theta)
            dt = L*np.cos(theta)
            plt.plot([t[i],t[i] + dt],[y[j],y[j] + dy],'b')
```

As an example, let's plot the slope field for the ODE $\frac{dy}{dt}=-ty$. 

```python
f = lambda t,y: -t*y
slopefield(f,[0,4],[-2,2],0.2,0.2)
```

Let's compare this to an analytic solution. We can use separation of variables to solve the ODE $\frac{dy}{dt}=-ty$ with initial condition $y(0)=C$, giving $y(t)=Ce^{-\frac{t^2}{2}}$. Let's begin with the choice of $C=1$. 

```python
C=1
y = lambda t: C*np.exp(-t**2/2)
t = np.linspace(0,4,200)
plt.plot(t,y(t),'r')
f = lambda t,y: -t*y
slopefield(f,[0,4],[-2,2],0.2,0.2)
```

The red curve starts at the initial condition $y(0)=1$. 

## Varying the Initial Condition

Now go and tweak the value of $C$ to start at different initial conditions $y(0)=C$. 

 

Notice that for $C>0$, the curve always slopes downward approaching a horizontal asymptote at $y=0$ while for $C<0$ it always slopes upwards. $C=0$ is the rather boring equation $y(t)=0$.
